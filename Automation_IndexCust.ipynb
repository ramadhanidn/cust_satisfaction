#Data Cleansing Customer Satisfaction Survey

```python
import pandas as pd
import os
import fnmatch
```

```python
# Return CSV File
for file in os.listdir():
    if fnmatch.fnmatch(file, '*.csv'):
        csv_file = file
        print(file)
sk = pd.read_csv(csv_file, sep=';')
```

```python
# See the Table
sk
```

```python
# See the column name in the Data Frame
idx_col = 0
print("Urutan Nomor Kolom dan Nama Kolom")
for i in sk.columns:
  print(idx_col,' ', i)
  idx_col += 1
```

```python
# Return Column with 'Very Dissatisfied', 'Dissatisfied', 'Satisfied', 'Very Satisfied'
matching_columns = sk.columns[sk.isin(['Very Dissatisfied', 'Dissatisfied', 'Satisfied', 'Very Satisfied']).any()]

# Change 'Very Dissatisfied': 1, 'Dissatisfied': 2, ''Satisfied': 3, 'Very Satisfied': 4
mapping = {'Very Dissatisfied': 1, 'Dissatisfied': 2, 'Satisfied': 3, 'Very Satisfied': 4}
sk_unsur = sk.loc[:, matching_columns].replace(mapping)

unsur = ['U' + str(i+1) for i, j in enumerate(matching_columns)]
unsur_dict = {key: value for key, value in zip(matching_columns, unsur)}

# Change name column into U1, U2, and so on
sk_unsur = sk_unsur.rename(columns = unsur_dict)  
```

```python
sk_unsur
```

```python
sk_clean = pd.merge(sk.iloc[:,0:8], sk_unsur, left_index=True, right_index=True)
sk_clean
```

```python
# Export data ke File excel

sk_clean.to_excel('skclean.xlsx', index = False)
```
